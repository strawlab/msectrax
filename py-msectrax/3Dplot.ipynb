```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
import glob
import os
from mpl_toolkits.mplot3d.axes3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable

#sort for latest file
list_of_files = glob.glob('*.csv') # * means all if need specific format then *.csv
latest_file = max(list_of_files, key=os.path.getctime)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx() 
plt.suptitle(latest_file, fontsize=16)
plt.ion()
fig.show()
fig.canvas.draw()

while True:

#     data=pd.read_csv(latest_file,skiprows=1,usecols=[0,1,2,3,4],header=None, names=['t','dac1','dac2','adc1','adc2'])
    data=pd.read_csv(latest_file,comment='#')

    data1=data
    data1=data1.astype(int)
    trw = data1.shape[0]
    data2=data1.iloc[trw-100:trw,:]
  
    ax1.clear()
    ax2.clear()
    ax1.set_ylabel('adc (V)')
    ax1.set_xlabel('Time (s)')
    s1=ax1.plot(data2.timestamp/1000,data2.adc1,'xkcd:lightblue',label = 'adc1')
    s2=ax1.plot(data2.timestamp/1000,data2.adc2,'xkcd:orange',label = 'adc2')
    ax1.minorticks_on()
    ax1.grid(which='major', linestyle='-', linewidth='0.5', color='black')
    ax1.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
      

    ax2.set_ylabel('dac (V)', color='blue')
    d1=ax2.plot(data2.timestamp/1000,data2.dac1,'xkcd:red',label = 'dac1')
    d2=ax2.plot(data2.timestamp/1000,data2.dac2,color=(0,0.1,0.8),label = 'dac2')
    ax2.tick_params(axis='y', labelcolor='blue')
    lns = s1+s2+d1+d2
    labs = [l.get_label() for l in lns]
    ax2.legend(lns, labs, loc='upper right', ncol=4)
    fig.tight_layout()
    fig.canvas.draw()
```
