```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
import glob
import os
from mpl_toolkits.mplot3d.axes3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable

#sort for latest file
list_of_files = glob.glob('*.csv') # * means all if need specific format then *.csv
latest_file = max(list_of_files, key=os.path.getctime)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx() 
plt.suptitle(latest_file, fontsize=16)
plt.ion()
fig.show()
fig.canvas.draw()

while True:
    data=pd.read_csv('2019-05-10T17.18.09.csv',comment='#')
#     data=pd.read_csv('2019-04-24T18.00.59.csv',skiprows=1,usecols=[0,1,2,3,4],header=None, names=['t','dac1','dac2','adc1','adc2'])
    data3a=pd.read_csv('2019-04-24T18.00.59.csv',skiprows=1,usecols=[0,1,2,3,4],header=None, names=['t','dac1','dac2','adc1','adc2'])
    data3=data3a.astype(float)
    data3.adc1=(1.5-data3.adc1*3.3/4096)/10
    data3.adc2=(1.5-data3.adc2*3.3/4096)/10
    data1=data.astype(float)
#     data1.adc1=(1.5-data1.adc1*3.3/4096)/10
#     data1.adc2=(1.5-data1.adc2*3.3/4096)/10
    trw = data1.shape[0]
    data2=data1.iloc[trw-100:trw,:]
  
    ax1.clear()
    ax2.clear()
    ax1.set_ylabel('adc (V)')
    ax1.set_xlabel('Time (s)')
    s1=ax1.plot(data2.t/1000,data2.adc1,'xkcd:lightblue',label = 'adc1')
    s2=ax1.plot(data2.t/1000,data2.adc2,'xkcd:orange',label = 'adc2')
    ax1.minorticks_on()
    ax1.grid(which='major', linestyle='-', linewidth='0.5', color='black')
    ax1.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
      

    ax2.set_ylabel('dac (V)', color='blue')
    d1=ax2.plot(data2.t/1000,data2.dac1,'xkcd:red',label = 'dac1')
    d2=ax2.plot(data2.t/1000,data2.dac2,color=(0,0.1,0.8),label = 'dac2')
    ax2.tick_params(axis='y', labelcolor='blue')
    lns = s1+s2+d1+d2
    labs = [l.get_label() for l in lns]
    ax2.legend(lns, labs, loc='upper right', ncol=4)
    fig.tight_layout()
    fig.canvas.draw()
```

```python
data4=data1.copy()
#data4.adc1=data4.adc1/data3.adc1
#data4.adc2=data4.adc2/data3.adc1
# data5 = data4[data4['adc1'] < 2]
# data5 = data5[data5['adc2'] < 2]
# data5 = data5[data5['adc2'] >-2]
# data5 = data5[data5['adc1'] >-2]
#print(data4.adc1)
data5 = data4[data4['dac1'] < 1500]
data5 = data5[data5['dac1'] > -2000]
data5 = data5[data5['dac2'] <2500]
data5 = data5[data5['dac2'] >-1000]

zero_adc1a=(data5[data5['dac2'] ==500])
zero_adc1b=(data5[data5['dac2'] ==1000])
zero_adc2a=(data5[data5['dac1'] ==-500])
zero_adc2b=(data5[data5['dac1'] ==0])


#data5.to_csv('central.csv')
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statistics as stat
import time
data6 = data5[data5['dac1'] == 0]#.iloc[1:9,:]
#print(data6)
x= np.asarray(data6['adc1'])
x1=x.reshape(x.shape[0],1)
y=np.asarray(data6['dac2'])
y1=y.reshape(y.shape[0],1)
dac2_model=LinearRegression()
dac2_model.fit(x1, y1)
dac2_predicted = dac2_model.predict(x1)
rmse = mean_squared_error(y1, dac2_predicted)
r2 = r2_score(y1,dac2_predicted)
dac2_slope=dac2_model.coef_
dac2_intercept=dac2_model.intercept_
print('dac2 function: dac2=',dac2_model.coef_,'*adc1+', dac2_model.intercept_,'---R2 score: ', r2 )

data7 = data5[data5['dac2'] == 1000]#.iloc[0:8,:]
x= np.asarray(data7['adc2'])
x1=x.reshape(x.shape[0],1)
y=np.asarray(data7['dac1'])
y1=y.reshape(y.shape[0],1)
dac1_model=LinearRegression()
dac1_model.fit(x1, y1)
dac1_predicted = dac1_model.predict(x1)
rmse = mean_squared_error(y1, dac1_predicted)
r2 = r2_score(y1,dac1_predicted)
dac1_slope=dac1_model.coef_
dac1_intercept=dac1_model.intercept_
print('dac1 function: dac1=',dac1_model.coef_,'*adc2+', dac1_model.intercept_,'---R2 score: ', r2 )


adc1_0=0.5*(stat.mean(zero_adc1a.adc1)+stat.mean(zero_adc1b.adc1))
adc2_0=0.5*(stat.mean(zero_adc2a.adc2)+stat.mean(zero_adc2b.adc2))
dac1_0=-250
dac2_0=750
curr_dac1=-5000
curr_dac2=-5000
temp_data=data1[data1['dac1']==curr_dac1]
temp_data=temp_data[temp_data['dac2']==curr_dac2]
curr_adc1=stat.mean(temp_data.adc1)
curr_adc2=stat.mean(temp_data.adc2)
#print('mean adc_0:',zero_adc1a, zero_adc1b)
print('mean adc_0:',adc1_0, adc2_0)
rs=[]



fig = plt.figure()
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx() 
#plt.suptitle(latest_file, fontsize=16)
plt.ion()
fig.show()
fig.canvas.draw()
rs.append([curr_dac1, curr_dac2,curr_adc1,curr_adc2])

while (1):
    

    P=0.5
    del_dac2=P*((dac2_slope[0][0]*curr_adc1+dac2_intercept[0])-dac2_0)

    del_dac1=P*((dac1_slope[0][0]*curr_adc2+dac1_intercept[0])-dac1_0)

        
    new_dac2=curr_dac2-del_dac2
    new_dac1=curr_dac1-del_dac1
    curr_dac2=new_dac2
    curr_dac1=new_dac1
#     print('curr_adc:',curr_adc1, curr_adc2)
#     print('Delta dacs:',del_dac1, del_dac2)
#     print('New dacs:',new_dac1, new_dac2)
    from random import *
    if abs(del_dac2) <0.001 and abs(del_dac2) <0.001:
        new_dac1= randint(-5000,  5000)
        new_dac2= randint(-5000,  5000)
    from numpy.polynomial import polynomial as P
    data8 = data1[data1['dac1'] == 0]
    x= np.asarray(data8['dac2'])
    x1=x.reshape(x.shape[0],1)
    y=np.asarray(data8['adc1'])
    y1=y.reshape(y.shape[0],1)
    c, stats = P.polyfit(x,y,20,full=True)
    #print(x)
    curr_adc1=P.polyval(new_dac2, c, tensor=True)
    #print(y_est)
    # fig = plt.figure()
    # plt.plot(x,y)
    # plt.plot(x,y_est)

    data9 = data1[data1['dac2'] == 1000]
    x= np.asarray(data9['dac1'])
    x1=x.reshape(x.shape[0],1)
    y=np.asarray(data9['adc2'])
    y1=y.reshape(y.shape[0],1)
    c, stats = P.polyfit(x,y,20,full=True)
    #print(x)
    curr_adc2=P.polyval(new_dac1, c, tensor=True)
    #print(dac1_0,dac2_0,adc1_0, adc2_0,new_dac1,new_dac2,curr_adc1,curr_adc2)
    print("dac1: %6.3f" % new_dac1,"dac2: %6.3f" % new_dac2,"adc1: %6.5f" % curr_adc1,"adc2: %6.5f" % curr_adc2)
    time.sleep(1)
    rs.append([new_dac1, new_dac2,curr_adc1,curr_adc2])
    rs1=np.asarray(rs)

    
    ax1.clear()
    ax2.clear()
    ax1.set_ylabel('DAC (V)')
    ax1.set_xlabel('Time (s)')
    s1=ax1.plot(rs1[:,0],'xkcd:lightblue',label = 'dac1')
    s2=ax1.plot(rs1[:,1],'xkcd:blue',label = 'dac2')
    ax1.minorticks_on()
    ax1.grid(which='major', linestyle='-', linewidth='0.5', color='black')
    ax1.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
    
    ax2.set_ylabel('ADC (V)')
    ax2.set_xlabel('Time (s)')
    s3=ax2.plot(rs1[:,2],'xkcd:orange',label = 'adc1')
    s4=ax2.plot(rs1[:,3],'xkcd:red',label = 'adc2')
     
    lns = s1+s2+s3+s4
    labs = [l.get_label() for l in lns]
    ax2.legend(lns, labs, loc='upper right', ncol=4)
    
    fig.tight_layout()
    fig.canvas.draw()
    #print(y_est)
    # fig = plt.figure()
    # plt.plot(x,y)
    # plt.plot(x,y_est)
    # fig = plt.figure()
    # plt.plot(stat.mean(zero_adc1.dac1),stat.mean(zero_adc1.adc1),'o')
    # plt.plot(zero_adc1.dac2,zero_adc1.adc2)
```
